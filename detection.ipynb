```python
# Import libraries
import pandas as pd
import numpy as np

# Load the CSV file (adjust filename if needed)
try:
    data = pd.read_csv('3ae033f50fa345051652.csv')  # Updated filename from document
    print("Data loaded successfully!")
    print(data.head())  # Check first few rows
except FileNotFoundError:
    print("File not found! Please check the filename or path.")
    exit()

# Calculate outliers for Percent Electric Vehicles
percent_ev = data['Percent Electric Vehicles']
Q1 = percent_ev.quantile(0.25)
Q3 = percent_ev.quantile(0.75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outliers
outliers = percent_ev[(percent_ev < lower_bound) | (percent_ev > upper_bound)]
outlier_rows = data.loc[data['Percent Electric Vehicles'].isin(outliers)]
print("Outliers in Percent Electric Vehicles with details:")
print(outlier_rows)

```
